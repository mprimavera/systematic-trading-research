```python
# Functions
# Lambda Expressions
# map and filter
# various useful methods
```

```python
def my_func(param = 1):
    """
    This is the docstring
    """
    return param * 5


my_func(3)
```

```python
def times_two(var):
    return var * 2

result = times_two(4)

result
```

```python
lambda var: var * 2
```

```python
seq = [1, 2, 3, 4, 5]
list(map(times_two, seq))
```

```python
seq_2 = [2, 4, 6, 8]
list(map(lambda num: num * 2, seq_2))
```

```python
def is_even(number):
    return number % 2 == 0

tuple(filter(is_even, seq))
```

```python
list(filter(lambda nums: (nums % 2 == 0), (10, 9, 8, 7, 6)))
```
