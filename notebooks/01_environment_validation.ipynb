```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("NumPy version:", np.__version__)
print("Pandas version:", pd.__version__)

np.random.seed(42)

returns = np.random.normal(0, 0.01, 252)
price = 100 * (1 + pd.Series(returns)).cumprod()

price.plot(title="Random Walk Simulation")
plt.show()
```
